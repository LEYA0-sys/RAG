<font color=grey size=50>进行相似度打分</font>

<font color=pink size=10>打分方法1</font>

1.BM25 + TF-IDF（余弦相似度）

2.对比简单检索的分数，观察提升后的检索方法是否有提高文本相似度

<font color=pink size=20>1.简单检索</font>

1.运行时间短，使用了整个QA对作为测试集，并且发现k从5-20-50变化的过程中，分数均明显增加

<font color=pink size=20>2.上下文压缩-llm</font>

1.主要采用k=50的方式计算BM25相似度和TF-IDF相似度分数

2.弊端：调用llm过滤不相关文档的话<font color =red>运行时间太慢</font>，只能随机抽取几个问题进行抽样检索

3.生成结果分析：与简单检索对比，<font color=red>BM25分数和TF-IDF分数均呈现提高趋势</font>


<font color=pink size=20>3.多查询检索-无提示词版本</font>

1.加载了QA对中所有问题进行问答

2.和简单检索相比avg_TFIDF_score明显提高，平均BM25相似度下降

3.与方法2-上下文压缩-llm相比，avg_TFIDF_score提高和平均BM25相似度提高



<font color=pink size=20>4.多查询检索-提示词版本</font>

1.加载了QA对中所有问题进行问答

2.和简单检索相比avg_TFIDF_score明显提高，平均BM25相似度下降

3.和无提示词版本对比，<font color=red>avg_TFIDF_score明显提高</font>



<font color=pink size=10>打分方法2</font>

1.比对检索到的页码的准确率和召回率

2.准确率计算的是检索到的正确页码占所有页码的比例

3.若检索到正确页码，召回率1，否则0

4.后续考虑添加评估指标：除了简单检索外，其他检索方式在检索到对应文档以后会进行处理再保存到输出结果中，考虑评估检索到的文档和答案来源文档的文本相似度 + 评估处理后的文档的质量

5.添加多查询+压缩的方式，缺点是压缩方法耗时过长

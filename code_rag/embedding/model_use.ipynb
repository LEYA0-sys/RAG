```python
from xinference.client import RESTfulClient

client = RESTfulClient("http://127.0.0.1:9997")
model_uid = "bge-large-zh-v1.5"  
model = client.get_model(model_uid)
texts = ["示例文本1", "示例文本2"]

# 创建嵌入
result = model.create_embedding(input=texts)

# 输出嵌入结果
for i, embedding in enumerate(result["data"]):
    print(f"文本 {i + 1} 的嵌入向量: {embedding['embedding']}")
```

```python
from xinference.client import RESTfulClient

client = RESTfulClient("http://127.0.0.1:9997")

# 查看当前已部署模型
print(client.list_models())

# 如果模型未部署，需要你先手动 deploy
uid = client.launch_model(
    model_name="bge-large-zh-v1.5",
    model_type="embedding",
    model_format="huggingface",
    quantization="none"  # 或 "q4" / "q8"，根据模型支持情况
)

print(f"模型已部署，UID: {uid}")
model = client.get_model(uid)
```
